Minibatch discrimination using sum of L1 distances
"Improved Techniques for Training GANs", Tim Salimans, Ian Goodfellow, Wojciech Zaremba, Vicki Cheung, Alec Radford, Xi Chen (https://arxiv.org/abs/1606.03498)

## Step 4: Define network functions

Our GAN uses some special functions that estimate how related each batch of data is to itself, and feed this information back into our model.

*sum_L1_distance_between_arrays()* takes the sum of all L1 distance between two arrays, and *sum_L1_distance_within_array()* takes this to compute the L1 distances between each element of an array and the rest of the array. 

This gives us the ability to measure how related two datasets are, and how much variance is within a single dataset.

*add_L1_col_to_variable()* is used by our generator to insert the within-array sum of L1 distances to the output of the generator, with a predefined amount of noise added to this amount.

*add noise to real()* does the same as *add_L1_col_to_variable()* to the real data, but also flips a proportion of the bits according to the *MAX_FLIP_THRESHOLD*. This will decrease over the course of training until it hits the *MIN_FLIP_THRESHOLD*. *result_noise* is then applied to the end result so that the real samples aren't just round numbers and look a bit more like the results of the generator (its dificult for the generator to produce very high or low numbers with the sigmoid function).

NB: many of these functions are not optimised for Pytorch - this would greatly improve the model.

```python
from sklearn.metrics.pairwise import manhattan_distances

def sum_L1_distance_between_arrays(array_1,array_2):
    return np.sum(manhattan_distances(array_1,array_2))

def sum_L1_distance_within_array(array):
    array_range = list(range(array.shape[0]))
    return sum([sum_L1_distance_between_arrays(array[n:n+1],np.delete(array,np.delete(array_range,array_range.index(3)),axis=0)) for n in array_range])
```

```python
MAX_L1_SUM = 50
L1_noise = 0.3

def add_L1_col_to_variable(data):
    L1_num = sum_L1_distance_within_array(data.data.cpu().numpy())
    L1_num = L1_num/MAX_L1_SUM
    
    L1_tensor = torch.FloatTensor(list(data.size())[0],1).zero_()
    
    for n in range(list(data.size())[0]):
        L1_tensor[n][0] = L1_num * random.uniform(1-L1_noise,1+L1_noise)
    
    L1_tensor = Variable(L1_tensor)
    if torch.cuda.is_available(): L1_tensor = L1_tensor.cuda()
        
#     print(torch.cat((data,L1_tensor),dim=1))
            
    return torch.cat((data,L1_tensor),dim=1)
```

```python
MAX_FLIP_THRESHOLD = 0.3
MIN_FLIP_THRESHOLD = 0.0
FLIP_THESHOLD = MAX_FLIP_THRESHOLD
result_noise = 0.02

def add_noise_to_real(data):
    
    L1_num = sum_L1_distance_within_array(data.numpy())
    L1_num = L1_num/MAX_L1_SUM
    
    L1_tensor = torch.FloatTensor(list(data.size())[0],1).zero_()
    
    for n in range(list(data.size())[0]):
        L1_tensor[n][0] = L1_num * random.uniform(1-L1_noise,1+L1_noise)
                      
    data = torch.cat((data,L1_tensor),dim=1)
    
    rows, cols = list(data.size())
    
    for a in range(rows):
        for b in range(cols-1):
            if random.uniform(0,1) < FLIP_THRESHOLD:
                if data[a][b] == 0.0:
                    data[a][b] = 1.0
                else:
                    data[a][b] = 0.0
            if data[a][b] == 0.0:
                data[a][b] = random.uniform(0,result_noise)
            elif data[a][b] == 1.0:
                data[a][b] = data[a][b] - random.uniform(0,result_noise)
    
    
#     for a in range(rows):
#         for b in range(cols):
#             data[a][b] = data[a][b]*random.uniform(1-noise,1+noise)
            
    return Variable(data)
```

## Step 5: Define network architechture

The Discriminator and Generator are defined below. Note the *add_L1_col_to_variable(x)* in the forward method of the Generator.
Smaller architechtures tended to generate better results, but this has not been comprehensively tested.

```python
###From original GAN implementation

class DiscriminatorNet(torch.nn.Module):
    """
    A three hidden-layer discriminative neural network
    """
    def __init__(self):
        super(DiscriminatorNet, self).__init__()
        n_features = input_neurons
        n_out = discriminator_output_neurons
        
        self.hidden0 = nn.Sequential( 
            nn.Linear(n_features, 32),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3)
        )
        self.hidden1 = nn.Sequential(
            nn.Linear(32, 16),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3)
        )
        self.out = nn.Sequential(
            torch.nn.Linear(16, n_out),
            torch.nn.Sigmoid()
        )

    def forward(self, x):
        x = self.hidden0(x)
        x = self.hidden1(x)
        x = self.out(x)
        return x
```

```python
###From original GAN implementation

class GeneratorNet(torch.nn.Module):
    """
    A three hidden-layer generative neural network
    """
    def __init__(self):
        super(GeneratorNet, self).__init__()
        n_features = generator_input_neurons
        n_out = input_neurons-1
        
        self.hidden0 = nn.Sequential(
            nn.Linear(n_features, 16),
            nn.LeakyReLU(0.2)
        )
        self.hidden1 = nn.Sequential(            
            nn.Linear(16, 32),
            nn.LeakyReLU(0.2)
        )
        self.out = nn.Sequential(
            nn.Linear(32, n_out),
            torch.nn.Sigmoid()
        )

    def forward(self, x):
        x = self.hidden0(x)
        x = self.hidden1(x)
        x = self.out(x)        
        x = add_L1_col_to_variable(x)
        
        return x
    
# Noise
def noise(size):
    n = Variable(torch.randn(size, generator_input_neurons))
    if torch.cuda.is_available(): return n.cuda() 
    return n
```

```python
###From original GAN implementation

discriminator = DiscriminatorNet()
generator = GeneratorNet()
if torch.cuda.is_available():
    discriminator.cuda()
    generator.cuda()
```

## Step 6: Define network training functions

The sequence of training for the two networks are defined below. The results of one become the loss of the other.

```python
###From original GAN implementation

def real_data_target(size):
    '''
    Tensor containing ones, with shape = size
    '''
    data = Variable(torch.ones(size, 1))
    if torch.cuda.is_available(): return data.cuda()
    return data

def fake_data_target(size):
    '''
    Tensor containing zeros, with shape = size
    '''
    data = Variable(torch.zeros(size, 1))
    if torch.cuda.is_available(): return data.cuda()
    return data
```

```python
###From original GAN implementation

def train_discriminator(optimizer, real_data, fake_data):
    # Reset gradients
    optimizer.zero_grad()
    
    # 1.1 Train on Real Data
    prediction_real = discriminator(real_data)
    # Calculate error and backpropagate
    error_real = loss(prediction_real, real_data_target(real_data.size(0)))
    error_real.backward()

    # 1.2 Train on Fake Data
    prediction_fake = discriminator(fake_data)
    # Calculate error and backpropagate
    error_fake = loss(prediction_fake, fake_data_target(real_data.size(0)))
    error_fake.backward()
    
    # 1.3 Update weights with gradients
    optimizer.step()
    
    # Return error
    return error_real + error_fake, prediction_real, prediction_fake

def train_generator(optimizer, fake_data):
    # 2. Train Generator
    # Reset gradients
    optimizer.zero_grad()
    # Generate prediction for our fake_data using discriminator
    prediction = discriminator(fake_data)
    # Calculate error and backpropagate
    error = loss(prediction, real_data_target(prediction.size(0)))
    error.backward()
    # Update weights with gradients
    optimizer.step()
    # Return error
    return error
```

```python
###From original GAN implementation

# generate samples for testing

num_test_samples = 16
test_noise = noise(num_test_samples)

#optimzer

d_optimizer = optim.Adam(discriminator.parameters(), lr=learning_rate)
g_optimizer = optim.Adam(generator.parameters(), lr=learning_rate)
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import pylab as pl

def kmeans_and_targets(fake_samples,real_samples,d_pred_fake,d_pred_real):

    synthetic_samples = fake_samples.data.cpu().numpy().round().astype(int)
    real_samples = real_data.data.cpu().numpy().round().astype(int)
    all_samples = np.concatenate((synthetic_samples,real_samples))
    targets = np.concatenate((np.zeros(len(synthetic_samples)),np.ones(len(real_samples))))

    pca = PCA(n_components=2).fit(all_samples)
    pca_2d = pca.transform(all_samples)
    pca_2d_synthetic = pca_2d[0:len(synthetic_samples)]
    pca_2d_real = pca_2d[len(synthetic_samples):]
    
    d_pred = np.concatenate((d_pred_fake.data.cpu().numpy(),d_pred_real.data.cpu().numpy())).round().astype(int)
               

    kmeans = KMeans(n_clusters=2,random_state=420).fit(all_samples)

    pl.figure('2 Cluster K-Means')
    pl.scatter(pca_2d[:, 0], pca_2d[:, 1], c=kmeans.labels_)
    pl.xlabel('Measure 1')
    pl.ylabel('Measure 2')
    pl.title('2 Cluster K-Means')
    pl.show()

    pl.figure('Discriminator Prediction')
    c1_count = 0
    c2_count = 0
    names = []
    for i in range(0, pca_2d.shape[0]):
        if d_pred[i] == 0:
            c1 = pl.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r')
            c1_count += 1
        else:
            c2 = pl.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g')
            c2_count += 1

    pl.xlabel('Measure 1')
    pl.ylabel('Measure 2')
    if c1_count > 0 and c2_count > 0:
        pl.legend([c1, c2], ['Predicted Fake', 'Predicted Real'])
    elif c1_count > 0:
        pl.legend([c1], ['Predicted Fake'])
    elif c2_count > 0:
        pl.legend([c2], ['Predicted Real'])
    pl.title('Discriminator Prediction')
    pl.show()

    pl.figure('Known Outcomes')
    c1 = pl.scatter(pca_2d_synthetic[:, 0], pca_2d_synthetic[:, 1], c='r',marker='+')
    c2 = pl.scatter(pca_2d_real[:, 0], pca_2d_real[:, 1], c='g',marker='*')
    pl.xlabel('Measure 1')
    pl.ylabel('Measure 2')
    pl.legend([c1, c2], ['Synthetic', 'Real'])
    pl.title('Known Outcomes')
    pl.show()
    
```

## Step 8: Train the network

I found it very important to have a variety of data about the GANs available as they trained. The following functions visualise the Generator's data and the Discriminator's predictions, the loss and accuracy of the nets, the L1 distance between the synthetic samples and real samples, and the L1 distance inside each set of samples, and the learning rate over time.

```python
###From original GAN implementation

logger = Logger(model_name='tabGAN', data_name='PETRODATA')

d_losses = []
d_accuracy = []
g_losses = []

L1_norm_real = []
L1_norm_fake = []
L1_norm_total = []

learning_rate_list = []

shakeup_freq = 1000 #how often you perform stochastic gradient descent with restarts
shakeup_magnitude = 10
SHAKEUP = False
CURRENT_LEARNING_RATE = learning_rate

for epoch in range(num_epochs):
    for n_batch, (real_batch,_) in enumerate(data_loader):
        
        # Apply Stochastic Gradient Descent with Restarts after 2 * shakeup_freq
        if epoch % shakeup_freq == 0 and epoch >= shakeup_freq*2:
            SHAKEUP = True
            
        if SHAKEUP:
            current_step = epoch % shakeup_freq
            CURRENT_LEARNING_RATE = learning_rate + (0.5*((learning_rate*shakeup_magnitude)-learning_rate)*
                                        (1+math.cos((current_step/shakeup_freq)*math.pi)))
            
            for d,g in zip(d_optimizer.param_groups,g_optimizer.param_groups):
                d['lr'] = CURRENT_LEARNING_RATE
                g['lr'] = CURRENT_LEARNING_RATE
                                
        #Adjust bit flip amount over time - the last few epochs will train on "true" data
        if epoch < 37000:
            FLIP_THRESHOLD = (((37000-epoch)/37000)*(MAX_FLIP_THRESHOLD-MIN_FLIP_THRESHOLD))+MIN_FLIP_THRESHOLD
        else:
            FLIP_THRESHOLD = 0

        ### Training  Discriminator
        
        # Apply L1, bit flip and noise transformations to the real data
        real_data = add_noise_to_real(real_batch)
        if torch.cuda.is_available(): real_data = real_data.cuda()          
        
        # Generate fake data
        fake_data = generator(noise(real_data.size(0))).detach()

        # Train Discriminator
        d_error, d_pred_real, d_pred_fake = train_discriminator(d_optimizer,
                                                                real_data, fake_data)

        ### Train Generator
        
        # Generate fake data
        fake_data = generator(noise(real_batch.size(0)))
        
        # Train Generator
        g_error = train_generator(g_optimizer, fake_data)
        
        # Log error
        logger.log(d_error, g_error, epoch, n_batch, num_batches)
        
        
        # Calculate discriminator accuracy
        real_size = d_pred_real.size(0)
        fake_size = d_pred_fake.size(0)
        total_test = real_size + fake_size
        correct_real = d_pred_real.data.cpu().numpy().round().sum()
        correct_fake = fake_size - d_pred_fake.data.cpu().numpy().round().sum()
        accuracy = (correct_real + correct_fake) / total_test
              
        # Record losses, accuracy and learning rate
        d_losses.append(d_error.data.cpu().numpy().tolist())
        g_losses.append(g_error.data.cpu().numpy().tolist())
        d_accuracy.append(accuracy)
        learning_rate_list.append(CURRENT_LEARNING_RATE)

        #Calculate and record L1 distances        
        fake_data_array = fake_data.data.cpu().numpy().round().astype(int)
        real_data_array = real_data.data.cpu().numpy().round().astype(int)

        L1_norm_total.append(sum_L1_distance_between_arrays(fake_data_array,real_data_array))
        L1_norm_fake.append(sum_L1_distance_within_array(fake_data_array))
        L1_norm_real.append(sum_L1_distance_within_array(real_data_array))
        
        #If L1 distances are greater than anticipated, then adjust the scaling of the variable
        
        if sum_L1_distance_within_array(fake_data_array) > MAX_L1_SUM: MAX_L1_SUM = sum_L1_distance_within_array(fake_data_array)
        if sum_L1_distance_within_array(real_data_array) > MAX_L1_SUM: MAX_L1_SUM = sum_L1_distance_within_array(real_data_array)
                              
        # Display Progress
        
        if epoch % 500 == 0:
                      
            display.clear_output(True)
            
            # General status
                
            logger.display_status(
                epoch, num_epochs, n_batch, num_batches,
                d_error.data.cpu().numpy().tolist(), g_error.data.cpu().numpy().tolist(), accuracy,
                d_pred_real.data.cpu().numpy(), d_pred_fake.data.cpu().numpy()
            )
            
            # Accuracy status
        
            print("correct real: {}/{}; correct fake: {}/{}".format(correct_real,real_size,
                                                                         correct_fake,fake_size))
            
            #Visualise datapoints
                        
            kmeans_and_targets(fake_data,real_data,d_pred_fake,d_pred_real)
            print()
            
            #Visualise the loss and accuracy over time
            
            plot_stats(d_losses,g_losses,d_accuracy)
            
            #Print L1 Distances
          
            print("Sum L1 distance between fake and real: {}".format(L1_norm_total[-1]))
            print("Sum L1 distance within real: {}".format(L1_norm_real[-1]))
            print("Sum L1 distance within fake: {}".format(L1_norm_fake[-1]))
            print()
            
            #Visualise L1 distances over time
            plot_L1_distances(L1_norm_real,L1_norm_fake,L1_norm_total)
            
            #Show learning rate
            plot_learning_rate(learning_rate_list)
            
            #Show actual examples for comparison (both raw and rounded as the samples will be)
            np.set_printoptions(threshold=sys.maxsize)
            print("sample real[0]")
            print(real_data.data.cpu().numpy()[0])
            print("sample fake[0]")
            print(fake_data.data.cpu().numpy()[0])
            
            print("rounded real[0]")
            print(real_data.data.cpu().numpy().round()[0])
            print("rounded fake[0]")
            print(fake_data.data.cpu().numpy().round()[0])
            
            #Generate synthetic data samples and save to the correct folder
            num = 100
            fake_data_array = generator(noise(num)).data.cpu().numpy().round().astype(int)
            target_array = np.full((num, 1), target_label)
            csv_array = np.concatenate((target_array,fake_data_array),axis=1)
            
            #Create directories if needed            
            filepath = Path("GAN_synthetic_samples/{}".format(target_label))
            os.makedirs(filepath, exist_ok=True)
            
            filename = Path(filepath,"GAN_synthetic_samples_target_{}_epoch_{}.csv".format(target_label,epoch))


            np.savetxt(filename, fake_data_array, delimiter=",")           
```

## Step 9: Dump stats to CSV for further analysis

The network's stats are saved in case they're needed to determine the best synthetic samples to use for training.

```python
import csv

#Create directories if needed            
filepath = Path("GAN_synthetic_samples/{}/stats".format(target_label))
os.makedirs(filepath, exist_ok=True)
filename = Path(filepath, "GAN_stats_target_{}.csv".format(target_label))

csv_body = [["Epoch","Discriminator Loss", "Generator Loss", "Discriminator Accuracy","Learning rate","Sum L1 Distance Real batch", "Sum L1 Distance Fake batch","Sum L1 Distance between batches"]]
epochs = range(num_epochs)

for a,b,c,d,e,f,g,h in zip(epochs,d_losses,g_losses,d_accuracy,learning_rate_list,L1_norm_real,L1_norm_fake,L1_norm_total):
    csv_body.append([a,b,c,d,e,f,g,h])

with open(filename, 'w',newline='') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerows(csv_body)

csvFile.close()
```
